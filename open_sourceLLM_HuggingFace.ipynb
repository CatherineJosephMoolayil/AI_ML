```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased", use_auth_token="")
```

```python
from transformers import pipeline

classifier = pipeline(task="text-classification", model="SamLowe/roberta-base-go_emotions", top_k=None)

sentences = ["While the sunset painted the sky with hues of orange and pink, John couldn't help but feel an overwhelming sense of melancholy and nostalgia, remembering the times he spent with his loved ones who were now miles away, and although he tried to focus on the beauty in front of him, the weight of loneliness settled heavily on his shoulders, making him yearn for the warmth of companionship and the laughter that once filled his evenings"]

model_outputs = classifier(sentences)

# Print each dictionary in a separate line
for output_dict in model_outputs:
    print(output_dict)


    # Print each label/score pair in a separate line
for result in model_outputs:
    for prediction in result:
        print(f"{prediction['label']}: {prediction['score']:.4f}")
```
